```python
#Application Task Scheduling in Cloud Computing

import random


num_tasks = 10
num_servers = 3
num_particles = 30
max_iterations = 50


tasks = [random.randint(1, 10) for _ in range(num_tasks)]


def fitness(particle):
    
    loads = [0] * num_servers
    for task_idx, server in enumerate(particle):
        loads[server] += tasks[task_idx]
    
    return max(loads)

def random_particle():
    return [random.randint(0, num_servers - 1) for _ in range(num_tasks)]

def optical_influence(particle_fitness):
    max_fit = 100  
    intensity = max(0, 1 - particle_fitness / max_fit)
    return intensity


swarm = [random_particle() for _ in range(num_particles)]
pbest = list(swarm)
pbest_fitness = [fitness(p) for p in pbest]


gbest_idx = pbest_fitness.index(min(pbest_fitness))
gbest = pbest[gbest_idx]
gbest_fitness = pbest_fitness[gbest_idx]

for iteration in range(max_iterations):
    for i, particle in enumerate(swarm):
       
        p_intensity = optical_influence(pbest_fitness[i])
        g_intensity = optical_influence(gbest_fitness)

        new_particle = particle.copy()

      
        for task_idx in range(num_tasks):
            if random.random() < p_intensity:
                
                new_particle[task_idx] = pbest[i][task_idx]
            if random.random() < g_intensity:
                
                new_particle[task_idx] = gbest[task_idx]

           
            if random.random() < 0.1:
                new_particle[task_idx] = random.randint(0, num_servers - 1)

        swarm[i] = new_particle
        current_fit = fitness(new_particle)

      
        if current_fit < pbest_fitness[i]:
            pbest[i] = new_particle
            pbest_fitness[i] = current_fit

           
            if current_fit < gbest_fitness:
                gbest = new_particle
                gbest_fitness = current_fit

    print(f"Iteration {iteration+1}, Best makespan: {gbest_fitness}")

print("\nBest task assignment to servers:")
for server in range(num_servers):
    assigned_tasks = [i for i, s in enumerate(gbest) if s == server]
    total_load = sum(tasks[i] for i in assigned_tasks)
    print(f" Server {server}: Tasks {assigned_tasks}, Total load = {total_load}")
```
